```python
# Loading in required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
personality_scores = pd.read_csv("personality_scores.csv", sep = ";")
personality_scores.head()
```

```python
personality_scores = personality_scores.dropna(axis = 'columns')
personality_scores.head()
```

```python
# Dropping Duplicates Based on ID
persona_scores = personality_scores[~personality_scores[['ID']].apply(frozenset, axis=1).duplicated()]
#persona_scores = personality_scores.drop_duplicates(subset ="ID", keep = False, inplace = True)

# Assert to check if new DF is length of original's DF entries
assert len(persona_scores) == len(personality_scores)
```

```python
# Creating a list of rows from the data
data_rows = personality_scores.values.tolist()

# Creating Function
def counting_personal_score_values(data_rows):
    # This function should return the total scores for each subscale, 
    # for each individual that took part in the test
    all_rows = []
    for row in data_rows:
        all_rows.append(row)
        
    # Creating lists that will contain the totals of scores for each subscale
    counters_table = []
    add_for_ones = []
    add_for_twos = []
    add_for_threes = []
    add_for_fours = []
    add_for_fives = []
    
    for each_row in all_rows:
        current_row = each_row
        # Creating counters for each subscale (wrt "scores")
        ones = 0
        twos = 0
        threes = 0
        fours = 0
        fives = 0
        
        for each_val in current_row:
            # Reading into each tuple individually and adding results into each counter, 
            # respectively
            if type(each_val) == str:
                if each_val[1] == '1':
                    ones += int(each_val[4])
                if each_val[1] == '2':
                    twos += int(each_val[4])
                if each_val[1] == '3':
                    threes += int(each_val[4])
                if each_val[1] == '4':
                    fours += int(each_val[4])
                if each_val[1] == '5':
                    fives += int(each_val[4])
        # Append calculated totals into list of each subscale, respectively
        add_for_ones.append(ones)
        add_for_twos.append(twos)
        add_for_threes.append(threes)
        add_for_fours.append(fours)
        add_for_fives.append(fives)
    # Append previously appended lists into a list (Creating a list of lists - Series)
    counters_table.append(add_for_ones)
    counters_table.append(add_for_twos)
    counters_table.append(add_for_threes)
    counters_table.append(add_for_fours)
    counters_table.append(add_for_fives)

    return counters_table
        
    


# Create a Dataframe from the Series
counters_table = counting_personal_score_values(data_rows)
df_count = pd.DataFrame(counters_table)
df_count = df_count.transpose()
# Rename columns accordingly
df_count = df_count.rename(columns={'0':'Extraversion',              
                                    '1':'Agreeableness', 
                                    '2':'Conscientiousness',
                                    '3':'Emotional Stability', 
                                    '4':'Intellect'})
df_count.columns = ['Extraversion',
                    'Agreeableness',
                    'Conscientiousness',
                    'Emotional Stability', 
                    'Intellect']
# df_count.shape
df_count.head()
```

```python
# personality_data = pd.merge(persona_scores, df_count, on = "ID", how = "inner")
personality_data = pd.concat([persona_scores, df_count], axis = 1, ignore_index = False)
personality_data.head()
```
