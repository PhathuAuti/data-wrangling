```python
data_rows = personality_scores.values.tolist()
def counting_personal_score_values(data_rows):
    # I want this function to count how many 1s, 2s, 3s, 4s and 5s appear per row
    rows = []
    for row in data_rows:
        rows.append(row)
        
    value_counts = []
    count_ones = []
    count_twos = []
    count_threes = []
    count_fours = []
    count_fives = []
    
    for each_row in rows:
        current_row = each_row
        ones = 0
        twos = 0
        threes = 0
        fours = 0
        fives = 0
        
        for each_val in current_row:

            if type(each_val) == str:
                if each_val[1] == '1':
                    ones += 1
                if each_val[1] == '2':
                    twos += 1
                if each_val[1] == '3':
                    threes += 1
                if each_val[1] == '4':
                    fours += 1
                if each_val[1] == '5':
                    fives += 1
        count_ones.append(ones)
        count_twos.append(twos)
        count_threes.append(threes)
        count_fours.append(fours)
        count_fives.append(fives)
    
    value_counts.append(count_ones)
    value_counts.append(count_twos)
    value_counts.append(count_threes)
    value_counts.append(count_fours)
    value_counts.append(count_fives)

    return value_counts
        
    


# Create rows to work with
value_counts = counting_personal_score_values(data_rows)
df_count = pd.DataFrame(value_counts)
df_count = df_count.transpose()
df_count.shape
df_count = df_count.rename(columns={
    '0':'Extraversion',              
    '1':'Agreeableness', 
    '2':'Conscientiousness',
    '3':'Neuroticism', 
    '4':'Intellect'})
df_count.columns = ['Extraversion',              
    'Agreeableness', 
    'Conscientiousness',
    'Neuroticism', 
    'Intellect']
df_count
# df_count.shape
```
