#### Create new columns containing the total score of each of the personality test subscales. 
#### Instructions:
To do this, write a function (or functions) that will calculate the total score for each of the subscales (conscientiousness, emotional stability, openness to new experience, agreeableness, extraversion), as set out in scoring. In other words, for the conscientiousness total score, all items marked as belonging to ‘conscientiousness’ should be summed.

```python
# Creating a list of rows from the data
data_rows = personality_scores.values.tolist()

# Creating Function
def counting_personal_score_values(data_rows):
    # This function should return the total scores for each subscale, 
    # for each individual that took part in the test
    all_rows = []
    for row in data_rows:
        all_rows.append(row)
        
    # Creating lists that will contain the totals of scores for each subscale
    sums_table = []
    add_for_ones = []
    add_for_twos = []
    add_for_threes = []
    add_for_fours = []
    add_for_fives = []
    
    for each_row in all_rows:
        current_row = each_row
        # Creating counters for each subscale (wrt "scores")
        ones = 0
        twos = 0
        threes = 0
        fours = 0
        fives = 0
        
        for each_val in current_row:
            # Reading into each tuple individually and adding results into each counter, 
            # respectively
            if type(each_val) == str:
                if each_val[1] == '1':
                    ones += int(each_val[4])
                if each_val[1] == '2':
                    twos += int(each_val[4])
                if each_val[1] == '3':
                    threes += int(each_val[4])
                if each_val[1] == '4':
                    fours += int(each_val[4])
                if each_val[1] == '5':
                    fives += int(each_val[4])
        # Append calculated totals into list of each subscale, respectively
        add_for_ones.append(ones)
        add_for_twos.append(twos)
        add_for_threes.append(threes)
        add_for_fours.append(fours)
        add_for_fives.append(fives)
    # Append previously appended lists into a list (Creating a list of lists - Series)
    sums_table.append(add_for_ones)
    sums_table.append(add_for_twos)
    sums_table.append(add_for_threes)
    sums_table.append(add_for_fours)
    sums_table.append(add_for_fives)

    return sums_table

# Create a Dataframe from the Series
sums_table = counting_personal_score_values(data_rows)
persona_sums = pd.DataFrame(sums_table)
persona_sums = persona_sums.transpose()
# Rename columns accordingly
persona_sums = persona_sums.rename(columns={'0':'Extraversion',              
                                    '1':'Agreeableness', 
                                    '2':'Conscientiousness',
                                    '3':'Emotional Stability', 
                                    '4':'Intellect'})
persona_sums.columns = ['Extraversion',
                    'Agreeableness',
                    'Conscientiousness',
                    'Emotional Stability', 
                    'Intellect']
# persona_sums.shape
persona_sums.head()
```

```python
# Concatenate Dataframe of Personality total scores with original Dataset
personality_data = pd.concat([persona_scores, persona_sums], axis = 1, ignore_index = False)
personality_data.head()
```

### Task 2

#### Import the data in departments.csv. 

```python
departments = pd.read_csv("departments.csv", sep = ";")
departments.shape
```

```python
departments.head()
```

#### Merge this data frame with the personality score data frame, keeping all applicants within the department data frame.

```python
# Merge two datasets, wrt ID column
per_dept = pd.merge(personality_data, departments, on = "ID", how = "inner")
per_dept.head()
```

#### Use an assert statement to check that the newly created merged data frame has the same amount of rows as the department data frame, and the expected number of columns.

```python
# Assert to check if new DF is length of original's DF entries, as well as columns
assert len(per_dept) == len(departments)
assert len(per_dept.columns) == 57
per_dept.shape
```

#### Filter the merged data frame so that you get only the applicants who scored less than 30 on emotional stability, conscientiousness AND agreeableness. 

```python
# Filter the people with scores less than 30 in three different columns.
df_filtered = per_dept[(per_dept['Agreeableness'] < 30) & (per_dept['Conscientiousness'] < 30) & (per_dept['Emotional Stability'] < 30)]
filtered1 = df_filtered[['ID', 'Department']]
filtered1
```

```python
# Filter the rest with scores higher than 30 (should add up to 1553 rows)
df_filtered = per_dept[(per_dept['Agreeableness'] >= 30) | (per_dept['Conscientiousness'] >= 30) | (per_dept['Emotional Stability'] >= 30)]
filtered2 = df_filtered[['ID', 'Department']]
filtered2
```

#### Further Instructions:
Print the ID numbers and departments of these applicants to the screen, and also assign these applicants the tag “high_risk” in a new column. All other applicants get the tag “low_risk”

```python
# Declare a list that is to be converted into a column in both datasets
filtered1['Risk'] = "high_risk"
filtered2['Risk'] = "low_risk"
```

```python
# Concatenate the two dataframes of filtered data, with the column 'Risk', sorting by ID
filtered_data = pd.concat([filtered1, filtered2]).sort_values(by = 'ID')
filtered_data
```
